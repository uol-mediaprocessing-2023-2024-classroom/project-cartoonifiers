```python
import cv2
from google.colab.patches import cv2_imshow

import numpy as np
```

```python
imgbase = cv2.imread('/content/drive/MyDrive/Colab Notebooks/OMO.jpeg')
grayimg = cv2.cvtColor(imgbase,cv2.COLOR_RGB2GRAY)
grayimg = cv2.medianBlur(grayimg, 5)
cv2_imshow(imgbase)
```

```python
edges = cv2.adaptiveThreshold(grayimg, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 9, 9)
cv2_imshow(edges)
```

```python
color = cv2.bilateralFilter(imgbase, 9, 250, 250)
cartoon = cv2.bitwise_and(color, color, mask=edges)
cv2_imshow(cartoon)
```
